```python
import csv

with open("學生成績.csv.csv",mode='r',encoding="utf-8",newline="") as file:
    csvReader=csv.reader(file)
    next(csvReader)
    #for row in csvReader:
    #    print(row)
    students = list(csvReader)
students
```

```python
print("國文>=80的學生有:")
amount = 0
for student in students:    
    if int(student[1]) >= 80:
        amount += 1
        print(f'學生:{student[0]},國文:{student[1]}')
print(f"共有{amount}學生")
```
